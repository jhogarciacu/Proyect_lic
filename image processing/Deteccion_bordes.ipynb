```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
ruta = "/content/drive/MyDrive/fotos_licorera/amarillo_1.jpeg"


imagen_1 = cv2.imread(ruta)

plt.imshow(cv2.cvtColor(imagen_1, cv2.COLOR_BGR2RGB))
plt.title("Imagen 1")
plt.axis("off")
plt.show()
```

```python
imagen_gris = cv2.cvtColor(imagen_1, cv2.COLOR_BGR2GRAY)

plt.imshow(imagen_gris, cmap="gray")
plt.title("Imagen en escala de grises")
plt.axis("off")
plt.show()
```

```python
suavizada = cv2.GaussianBlur(imagen_gris, (5, 5), 0)

plt.imshow(suavizada, cmap="gray")
plt.title("Imagen suavizada")
plt.axis("off")
plt.show()
```

```python
imgen_bordes = cv2.Canny(suavizada, threshold1=100, threshold2=200)

plt.imshow(imgen_bordes, cmap="gray")
plt.title("Imagen con bordes")
plt.axis("off")
plt.show()
```

```python
kernel = np.ones((3,3), np.uint8)
bordes_refinados = cv2.morphologyEx(imgen_bordes, cv2.MORPH_CLOSE, kernel)

plt.imshow(bordes_refinados, cmap="gray")
plt.title("Imagen con bordes")
plt.axis("off")
plt.show()
```

```python
contornos, _ = cv2.findContours(bordes_refinados, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contornos = [c for c in contornos if cv2.contourArea(c) > 500]
resultado = imagen_1.copy()
cv2.drawContours(resultado, contornos, -1, (0,255,0), 2)

plt.imshow(resultado, cmap="gray")
plt.title("Imagen con bordes")
plt.axis("off")
plt.show()
```
